```python
import pandas as pd
import numpy as np
from datetime import datetime
```

Load Complete data

```python
df = pd.read_csv('melb_data.csv')
num_columns = len(df.columns)
pd.set_option("display.max_columns", num_columns)

feature_names = df.columns.tolist()
feature_names
```

Show the types of Variables

```python
df.info()
```

Checking the unique values to see for mispells

```python
df['Car'].unique()
```

Removing Duplicates

```python
df.drop_duplicates(keep=False,inplace=True)
df
```

Replace NaNs with default values

```python
df['Car'] = df['Car'].replace(np.nan, 1)
df['Car'].unique()

df['BuildingArea'] = df['BuildingArea'].replace(np.nan, 80.0)
df['BuildingArea'].unique()

df['YearBuilt'] = df['YearBuilt'].replace(np.nan, 2000)
df['YearBuilt'].unique()
```

```python
grp = df.groupby(['Suburb'])
grp.describe()
```
