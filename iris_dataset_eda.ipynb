Univariate Analysis

```python
sns.FacetGrid(df,hue="Species",height=5).map(sns.distplot,"PetalWidthCm")
```

```python
sns.FacetGrid(df,hue="Species",height=5).map(sns.distplot,"PetalLengthCm")
```

```python
sns.FacetGrid(df,hue="Species",height=5).map(sns.distplot,"SepalWidthCm")
```

```python
sns.FacetGrid(df,hue="Species",height=5).map(sns.distplot,"SepalLengthCm")
```

```python
sns.violinplot(x="Species",y="PetalLengthCm", data=df, size=8)
plt.show()
```
